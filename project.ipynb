```python
#importing libraries
import pandas as pd
import seaborn as sns
from datasets import load_dataset
import matplotlib.pyplot as plt
import ast

#loading data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

#data cleanup
df['job_posted_date']=pd.to_datetime(df.job_posted_date)
df['job_skills'] = df['job_skills'].apply(lambda skill_list: ast.literal_eval(skill_list) if isinstance(skill_list, str) else skill_list)
```

```python
df_da_eg = df[(df['job_country'] == 'Egypt') & (df['job_title_short'] == 'Data Analyst')]
```

```python
df_da_eg
```

```python
# Count occurrences of each job_location and create a DataFrame  
df_plot = df_da_eg['job_location'].value_counts().head(10).reset_index()  
df_plot.columns = ['job_location', 'count']  

# Create a bar plot using Seaborn  
sns.set_theme(style='ticks')
sns.barplot(data=df_plot, x='count', y='job_location', palette='dark:b')
sns.despine()
plt.title('counts of job locations for data analyst in Egypt')
plt.xlabel('Number of jobs')
plt.ylabel('')
plt.show()
```

```python
df_plot = df_da_eg['company_name'].value_counts().to_frame().head(10).reset_index()

sns.barplot(data=df_plot, x='count', y='company_name', palette='dark:b')
sns.set_theme(style='ticks')
sns.despine()
plt.title('counts of job / company for data analyst in Egypt')
plt.xlabel('Number of jobs')
plt.ylabel('')
plt.show()
```
